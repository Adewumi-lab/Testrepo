##### Import Necessary Modules:


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.
</p>


The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
# Display basic info about the DataFrame (columns, types, non-null counts)
print("\nDataFrame Info:")
print(df.info())

# Display summary statistics for numeric columns
print("\nSummary Statistics:")
print(df.describe())

# Optional: Summary stats for non-numeric columns (like counts, unique values)
print("\nSummary for object columns:")
print(df.describe(include=['object']))
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Count missing values per column
missing_counts = df.isnull().sum()

print("Missing values per column:")
print(missing_counts)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in Dataset')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_mainbranch = df['MainBranch'].isnull().sum()
print(f"Number of missing rows in 'MainBranch': {missing_mainbranch}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
most_frequent_mainbranch = df['MainBranch'].mode()[0]
print(f"The most frequent value in 'MainBranch' is: {most_frequent_mainbranch}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Find the most frequent value (mode)
most_frequent = df['MainBranch'].mode()[0]

# Fill missing values with the mode
df['MainBranch'].fillna(most_frequent, inplace=True)

# Verify no missing values remain
missing_after = df['MainBranch'].isnull().sum()
print(f"Missing values in 'Emp' after imputation: {missing_after}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Count occurrences of each category in Employment
employment_counts = df['Employment'].value_counts()

# Plot the distribution
plt.figure(figsize=(10,6))
employment_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Employment Status')
plt.xlabel('Employment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout(rect=[0, 0.1, 1, 1])  # leave more space at bottom
plt.show()
```
