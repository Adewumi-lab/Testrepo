## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# Count the number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Get the duplicate rows themselves
duplicate_rows = df[df.duplicated()]

# Display the first few duplicate rows
print("First few duplicate rows:")
print(duplicate_rows.head())
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# Step 1: Find duplicates based on subset of columns
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']
duplicates_subset = df[df.duplicated(subset=subset_cols, keep=False)]  # keep=False marks all duplicates

print(f"Number of duplicates based on {subset_cols}: {len(duplicates_subset)}")

# Step 2: Group duplicates by the subset columns
grouped = duplicates_subset.groupby(subset_cols)

# Step 3: For each group, check which other columns have identical values in all rows
consistent_cols = []

for col in df.columns:
    if col in subset_cols:
        continue  # skip subset columns, already grouped by these
    # Check if values are the same in each group for this column
    # For each group, test if unique values count == 1
    all_groups_consistent = grouped[col].nunique() == 1
    if all_groups_consistent.all():
        consistent_cols.append(col)

print("Columns that have identical values across all duplicates grouped by subset columns:")
print(consistent_cols)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
import matplotlib.pyplot as plt
# Identify duplicates (based on all columns)
duplicates = df[df.duplicated(keep=False)]

# Count duplicates by Country
country_counts = duplicates['Country'].value_counts()

# Bar chart for Country
plt.figure(figsize=(10, 5))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Duplicate Rows by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Count duplicates by Employment
employment_counts = duplicates['Employment'].value_counts()

# Pie chart for Employment
plt.figure(figsize=(7, 7))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Duplicate Rows by Employment')
plt.axis('equal')
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
critical_cols = ['MainBranch', 'Employment', 'RemoteWork']  # example columns
df_unique = df.drop_duplicates(subset=critical_cols, keep='first')
print(f"Original rows: {len(df)}")
print(f"Rows after removing duplicates based on {critical_cols}: {len(df_unique)}")
num_removed = len(df) - len(df_unique)
print(f"Number of duplicates removed: {num_removed}")
```
